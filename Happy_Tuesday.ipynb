#### Happy Tuesday is a code let operator feel more intersting in their boring Tuesday job by Automation script
單上面的圖片會在資料夾中以10秒一次的速度轉換，而順序也會依照當週排序而定
用Beautiful soup處理要抓的網址，同時用chromedriver 和 screenshot來擷取目標的圖
下方也有可以同步抓取的目標圖片名稱取名為檔名的版本，速度會稍稍降低，但目前以圖片認定及可以
效益：預期節省每週4小時的人力成本
